```python
import numpy as np
import pandas as pd
```

```python
Books=pd.read_csv("/content/drive/MyDrive/movie/Books.csv")
Users=pd.read_csv("/content/drive/MyDrive/movie/Users.csv")
Rating=pd.read_csv("/content/drive/MyDrive/movie/Ratings.csv")
```

```python
print(Books.shape)
print(Users.shape)
print(Rating.shape)
```

```python
Books_with_rating=Rating.merge(Books,on="ISBN")
```

```python
Books_with_rating
```

```python
num_bookrating_df=Books_with_rating.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_bookrating_df.rename(columns={'Book-Rating':'num_Book_ratings'},inplace=True)
num_bookrating_df
```

```python
avg_bookrating_df=Books_with_rating.groupby("Book-Title").mean()["Book-Rating"].reset_index()
avg_bookrating_df.rename(columns={'Book-Rating':'avg_Book_ratings'},inplace=True)
avg_bookrating_df
```

```python
popular_Books_df = num_bookrating_df.merge(avg_bookrating_df,on='Book-Title')
popular_Books_df
```

```python
popular_Books_df[popular_Books_df["num_Book_ratings"]>200].sort_values("avg_Book_ratings",ascending=False).head(100)
```

```python
popular_Books_df=popular_Books_df.merge(Books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title',"Book-Author",'Image-URL-M','num_Book_ratings',"avg_Book_ratings"]]
```

```python
popular_Books_df
```

```python
x=Books_with_rating.groupby("User-ID").count()["Book-Rating"]>150
rated_user=x[x].index
```

```python
filtered_user_rating=Books_with_rating[Books_with_rating["User-ID"].isin(rated_user)]
```

```python
y=filtered_user_rating.groupby("Book-Title").count()["Book-Rating"]>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_user_rating[filtered_user_rating["Book-Title"].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt=final_ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:11]
    for i in similar_items:
      print(pt.index[i[0]])
```
